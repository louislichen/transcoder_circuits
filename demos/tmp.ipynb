```python
import numpy as np
import matplotlib.pyplot as plt

def f(gamma, alpha):
    return gamma - (1 - alpha**(gamma + 1)) / (1 - alpha)

alpha = np.linspace(0.01, 0.99, 500)

plt.figure(figsize=(8, 6))
for gamma in range(1, 11):
    y = f(gamma, alpha)
    plt.plot(alpha, y, label=f' = {gamma}')

plt.xlabel('', fontsize=12)
plt.ylabel(r'$\gamma - \frac{1-\alpha^{\gamma+1}}{1-\alpha}$', fontsize=14)
plt.title('Function curves for different  values')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
